```python
def decisionStump(df, w):
    
    
    data = df.loc[:, df.columns != 'label']
    lab = df['label']
    minerror = 1000
    signs = [ 'less', 'greater']
    
    for column in data:
        sliced = df[column]
        for i in range(len(sliced)):
            for sign in signs:
                if(sign=='less'):
                    
                    if(type(sliced)==type(pd.Series())):
                        sliced = sliced.as_matrix()
                        predicted = np.where(sliced < sliced[i], -1, 1)
                    else:
                        predicted = np.where(sliced < sliced[i], -1, 1)
                elif (sign == 'greater'):
                    if(type(sliced)==type(pd.Series())):
                        sliced = sliced.as_matrix()
                        predicted = np.where(sliced > sliced[i], -1, 1)
                    else:
                        predicted = np.where(sliced > sliced[i], -1, 1)

                error = errorRate(predicted, lab, w, len(lab.loc[lab==-1]), len(lab.loc[lab==1]))
                if(error<minerror):
                    minerror = error
                    optFeature = column
                    optLab = predicted
                    optThresh = sliced[i]
                    optSign = sign
    return optFeature, optLab, optThresh, minerror, optSign

def decisionStumpFast(df, w):
    
    
    data = df.loc[:, df.columns != 'label']
    lab = df['label']
    minerror = 1000
    signs = [ 'less', 'greater']
    
    for column in data:
        sliced = df[column]
        thresholds = np.linspace(min(sliced), max(sliced), 20)
        for i in range(len(thresholds)):
            for sign in signs:
                if(sign=='less'):
                    
                    if(type(sliced)==type(pd.Series())):
                        sliced = sliced.as_matrix()
                        predicted = np.where(sliced < thresholds[i], -1, 1)
                    else:
                        predicted = np.where(sliced < thresholds[i], -1, 1)
                elif (sign == 'greater'):
                    if(type(sliced)==type(pd.Series())):
                        sliced = sliced.as_matrix()
                        predicted = np.where(sliced > thresholds[i], -1, 1)
                    else:
                        predicted = np.where(sliced > thresholds[i], -1, 1)

                error = errorRate(predicted, lab, w, len(lab.loc[lab==-1]), len(lab.loc[lab==1]))
                if(error<minerror):
                    minerror = error
                    optFeature = column
                    optLab = predicted
                    optThresh = thresholds[i]
                    optSign = sign
    return optFeature, optLab, optThresh, minerror, optSign
        
        
def errorRate(predicted, true, w, obj1, obj2, prior1 = 0.5, prior2 = 0.5):
    
#     mis1 = np.sum(predicted[:obj1] != true[:obj1])
#     mis2 = np.sum(predicted[obj1:] != true[obj1:])
    
#     return (prior1*mis1/obj1 + prior2*mis2/obj2)
    return np.sum(w*(predicted != true))

def createData(point):
        a = np.random.multivariate_normal([0, 0], [[1,0],[0,1]], point)
        b = np.random.multivariate_normal([2, 0], [[1,0],[0,1]], point)
        df = pd.DataFrame(np.concatenate([a,b]), columns =[ 'x', 'y'])
        lab0 = -np.ones(point)
        lab1 = np.ones(point)


        df['label'] = np.append(lab0, lab1)
        return df
    
def trainTest(digit, train, zerosize,  onesize ):
    zero = digit[:zerosize]
    one = digit[zerosize:]
    train0 = zero.sample(frac=train/zerosize)
    test0 = zero.drop(train0.index)
    train1 = one.sample(frac = train/onesize)
    test1 = one.drop(train1.index)
    return train0, test0, train1, test1

def trueError(dftst,f, t, s):
    slicedtst = dftst[f]
    slicedtst = slicedtst.as_matrix()
    if(s == 'less'):
        predicted = np.where(slicedtst < t, -1, 1)
    elif(s=='greater'):
        predicted = np.where(slicedtst > t, -1, 1)
    return predicted
#     return errorRate(predicted, lab , obj1 = len(lab.loc[lab==0]), obj2 = len(lab.loc[lab==1]))

# def weightedError():
        
```

```python
def calculateAlpha(w, e):
    return 0.5*np.log((sum(w)/(e)) -1)

def updateWeight(sumWeight, label):
    
    y = np.multiply(label,sumWeight)
    return np.exp(-y)
```

```python

def adaBoost(df, K):
    w = np.ones(len(df))
    alpha = []
    sumWeight = 0
    threshold = []
    feature = []
    sign = []
    weak = []
    label = df['label']
    for i in range(K):
#         print(i)    
        f,l,t,e,s  = decisionStumpFast(df, w)
#         print(e)
#         print(sum(w))
        if(e == 0):
            e = 0.0001
        weak.append(l)
#         print(weak)
        feature.append(f)
        threshold.append(t)
        sign.append(s)
#         print(weak)
        alpha.append(calculateAlpha(w, e))
#         print(alpha)
        sumWeight+=alpha[i]*weak[i]
        w = updateWeight(sumWeight, label)
#         print(w)
    return feature, threshold, sign, alpha
```

```python
def adaBoostAccuracy(f, t, s, alpha, test, K):
    lab = test[ 'label']
    discriminator = np.zeros(len(test))
    for i in range(K):
        discriminator = discriminator + alpha[i]*trueError(test, f[i], t[i], s[i] )
    predicted = np.where(discriminator > 0, 1, -1)
    return errorRate(predicted, lab , np.ones(len(test)),len(lab.loc[lab==-1]), len(lab.loc[lab==1]))
```

```python
acc = []
data = createData(200)
for round1 in range(200):
    print(round1)
    
    train0, test0, train1, test1 = trainTest(data, 50, 200, 200  )
    df =  pd.concat([train0, train1])
    feature, threshold, sign , alpha = adaBoost(df, 10)
    dftst = pd.concat([test0, test1])
    lab = dftst['label']
    print(adaBoostAccuracy(feature, threshold, sign, alpha, dftst, 10)/len(dftst))
    acc.append(adaBoostAccuracy(feature, threshold, sign, alpha, dftst, 10)/len(dftst))
print('True error: \t Mean {} \t Standard Deviation {}'.format(np.mean(acc), np.std(acc)))
```
